```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler

from sklearn.cluster import KMeans

# Load data
data = pd.read_csv('C:\\Users\\GEETHU\\Documents\\New folder\\train_u6lujuX_CVtuZ9i.csv')

```

```python
# import necessary libraries
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

data.isnull().sum()
data.isnull().sum()*100 / len(data)

columns = ['Gender','Dependents','LoanAmount','Loan_Amount_Term']
data = data.dropna(subset=columns)
data.isnull().sum()*100 / len(data)
data['Self_Employed'].mode()[0]
data['Self_Employed'] =data['Self_Employed'].fillna(data['Self_Employed'].mode()[0])
data.isnull().sum()*100 / len(data)
data['Gender'].unique()
data['Self_Employed'].unique()
data['Credit_History'].mode()[0]
data['Credit_History'] =data['Credit_History'].fillna(data['Credit_History'].mode()[0])
data.isnull().sum()*100 / len(data)
data['Dependents'] =data['Dependents'].replace(to_replace="3+",value='4')
data['Dependents'].unique()
data['Loan_Status'].unique()


# prepare data for clustering
approved = data[data['Loan_Status']=='Y'][['ApplicantIncome', 'Credit_History', 'LoanAmount']]
unapproved = data[data['Loan_Status']=='N'][['ApplicantIncome', 'Credit_History', 'LoanAmount']]

# perform clustering on approved applicants
kmeans_app = KMeans(n_clusters=3, random_state=0).fit(approved)

# perform clustering on unapproved applicants
kmeans_unapp = KMeans(n_clusters=3, random_state=0).fit(unapproved)

# visualize clustering results
fig, axs = plt.subplots(2, 3, figsize=(15,10))
axs[0,0].scatter(approved['ApplicantIncome'], approved['Credit_History'], c=kmeans_app.labels_)
axs[0,0].set_title('Approved Applicants')
axs[0,0].set_xlabel('Applicant Income')
axs[0,0].set_ylabel('Credit History')

axs[0,1].scatter(approved['ApplicantIncome'], approved['LoanAmount'], c=kmeans_app.labels_)
axs[0,1].set_title('Approved Applicants')
axs[0,1].set_xlabel('Applicant Income')
axs[0,1].set_ylabel('Loan Amount')

axs[0,2].scatter(approved['Credit_History'], approved['LoanAmount'], c=kmeans_app.labels_)
axs[0,2].set_title('Approved Applicants')
axs[0,2].set_xlabel('Credit History')
axs[0,2].set_ylabel('Loan Amount')

axs[1,0].scatter(unapproved['ApplicantIncome'], unapproved['Credit_History'], c=kmeans_unapp.labels_)
axs[1,0].set_title('Unapproved Applicants')
axs[1,0].set_xlabel('Applicant Income')
axs[1,0].set_ylabel('Credit History')

axs[1,1].scatter(unapproved['ApplicantIncome'], unapproved['LoanAmount'], c=kmeans_unapp.labels_)
axs[1,1].set_title('Unapproved Applicants')
axs[1,1].set_xlabel('Applicant Income')
axs[1,1].set_ylabel('Loan Amount')

axs[1,2].scatter(unapproved['Credit_History'], unapproved['LoanAmount'], c=kmeans_unapp.labels_)
axs[1,2].set_title('Unapproved Applicants')
axs[1,2].set_xlabel('Credit History')
axs[1,2].set_ylabel('Loan Amount')

plt.show()
```

```python
# import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import KMeans



# prepare data for clustering
approved_customers = data[data['Loan_Status']=='Y']
unapproved_customers = data[data['Loan_Status']=='N']

X_approved = approved_customers[['ApplicantIncome', 'Credit_History', 'LoanAmount']]
X_unapproved = unapproved_customers[['ApplicantIncome', 'Credit_History', 'LoanAmount']]

# perform clustering using KMeans
kmeans_approved = KMeans(n_clusters=3, random_state=0).fit(X_approved)
kmeans_unapproved = KMeans(n_clusters=3, random_state=0).fit(X_unapproved)

# visualize the clusters
fig = plt.figure(figsize=(12,6))
ax1 = fig.add_subplot(121, projection='3d')
ax1.scatter(X_approved['ApplicantIncome'], X_approved['Credit_History'], X_approved['LoanAmount'], c=kmeans_approved.labels_, cmap='viridis')
ax1.scatter(kmeans_approved.cluster_centers_[:,0], kmeans_approved.cluster_centers_[:,1], kmeans_approved.cluster_centers_[:,2], s=200, alpha=0.5, c='red')
ax1.set_title('Approved Customers Clusters')
ax1.set_xlabel('Applicant Income')
ax1.set_ylabel('Credit History')
ax1.set_zlabel('Loan Amount')


ax2 = fig.add_subplot(122, projection='3d')
ax2.scatter(X_unapproved['ApplicantIncome'], X_unapproved['Credit_History'], X_unapproved['LoanAmount'], c=kmeans_unapproved.labels_, cmap='viridis')
ax2.scatter(kmeans_unapproved.cluster_centers_[:,0], kmeans_unapproved.cluster_centers_[:,1], kmeans_unapproved.cluster_centers_[:,2], s=200, alpha=0.5, c='red')
ax2.set_title('Unapproved Customers Clusters')
ax2.set_xlabel('Applicant Income')
ax2.set_ylabel('Credit History')
ax2.set_zlabel('Loan Amount')

plt.show()
label1 = kmeans_approved.fit_predict(X_approved)
print(label1)
label2 = kmeans_unapproved.fit_predict(X_unapproved)
print(label2)


print(kmeans_approved.cluster_centers_)
print(kmeans_unapproved.cluster_centers_)
```

```python
from sklearn.metrics import silhouette_score

# calculate the silhouette score for approved customers
silhouette_approved = silhouette_score(X_approved, kmeans_approved.labels_)
print("Silhouette Score for Approved Customers:", silhouette_approved)

# calculate the silhouette score for unapproved customers
silhouette_unapproved = silhouette_score(X_unapproved, kmeans_unapproved.labels_)
print("Silhouette Score for Unapproved Customers:", silhouette_unapproved)
```

```python
# predict cluster for new customer
new_customer = [[15000, 0, 5000]] # example data for new customer
if new_customer[0][1] == 1:
    cluster = kmeans_approved.predict(new_customer)
    print(f'The new customer belongs to approved cluster {cluster[0]}')
else:
    cluster = kmeans_unapproved.predict(new_customer)
    print(f'The new customer belongs to unapproved cluster {cluster[0]}')
```
